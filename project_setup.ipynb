```python
import os

base = "thesis_assistant"

# Function to remove the existing thesis_assistant directory
def remove_thesis_assistant(base_path):
    if os.path.exists(base_path):
        for root, dirs, files in os.walk(base_path, topdown=False):
            for name in files:
                os.remove(os.path.join(root, name))
            for name in dirs:
                os.rmdir(os.path.join(root, name))
        os.rmdir(base_path)
        print(f"Removed existing directory: {base_path}")

# Remove the existing directory structure to start fresh
remove_thesis_assistant(base)


# Step 1: Define all directories (relative to thesis_assistant/)
directories = {
    "ethics": ["ethics_module.py"],
    "writing": ["writing_support.py"],
    "brainstorming": ["idea_brainstorming.py"],
    "emotion": ["emotion_support.py"],
    "rl_agent": ["ethics_supervisor_rl.py"],
    "langgraph_flow": ["nodes.py", "flow.py"],
    "orchestrator": ["thesis_assistant.py"],
}

# Step 2: Create folders and empty files
base = "thesis_assistant"
for folder, files in directories.items():
    dir_path = os.path.join(base, folder)
    os.makedirs(dir_path, exist_ok=True)
    for file in files:
        file_path = os.path.join(dir_path, file)
        if not os.path.exists(file_path):
            with open(file_path, "w") as f:
                f.write(f"# {file} — autogenerated module\n")

print("Directory + module files created successfully.")

def print_ordered_tree(base, directories):
    print(f"{base}/")
    for folder, files in directories.items():
        parts = folder.split('/')
        indent = '│   ' * len(parts)
        print(f"{indent}├── {parts[-1]}/")
        for file in files:
            print(f"{indent}│   └── {file}")

print_ordered_tree("thesis_assistant",directories)
```
